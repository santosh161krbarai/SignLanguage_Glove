```python
import sklearn
print(sklearn.__version__)
```

```python
import pandas as pd
```

```python
df=pd.read_csv("sensor_data_alpha_\"O\"_250_new_sensors.csv")
```

```python
df.head()
```

```python
df['label'] = df['label'].replace({'O': 0, 'C': 1})
```

```python
df.head()
```

```python
df.tail()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(df.drop(['label'],axis='columns'),df['label'],test_size=0.3)
```

```python
len(x_train)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predicted=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sm
import numpy as np
plt.figure(figsize=(5,5))
sm.heatmap(cm,annot=True)
plt.xlabel=('predicted')
plt.ylabel=('Truth')
```

```python
import joblib

joblib.dump(model, "my_model_newSensorData.joblib")
print("Model saved as my_modelnew.joblib")
```
